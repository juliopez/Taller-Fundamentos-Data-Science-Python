
## ‚è±Ô∏è (8‚Äì10 min) Actividad 3 ‚Äî Filtrar datos
**Objetivo:** Practicar condiciones l√≥gicas y selecci√≥n de columnas.

**Preguntas de gu√≠a:**
- ¬øCu√°ntos estudiantes reprobaron?
- ¬øCu√°l es la distribuci√≥n de edades en la carrera de Educaci√≥n?


```python

# Estudiantes que reprobaron
reprobados = df[df["Aprobado"] == False]
print("Total reprobados:", reprobados.shape[0])
reprobados.head()
```

```python

# Estudiantes de Educaci√≥n
edu = df[df["Carrera"] == "Educaci√≥n"]
print("Estudiantes de Educaci√≥n:", edu.shape[0])
edu["Edad"].describe()
```


## ‚è±Ô∏è (10‚Äì12 min) Actividad 4 ‚Äî Agrupar y resumir
**Objetivo:** Usar `groupby` para an√°lisis simple.

**Preguntas de gu√≠a:**
- ¬øQu√© carrera tiene el promedio de nota m√°s alto?
- ¬øCu√°l carrera tiene la asistencia promedio m√°s baja?


```python

promedios_nota = df.groupby("Carrera")["Nota_Final"].mean().sort_values(ascending=False)
promedios_asistencia = df.groupby("Carrera")["Asistencia_%"].mean().sort_values()

print("Promedio de Nota por Carrera (desc):")
display(promedios_nota)

print("\nPromedio de Asistencia por Carrera (asc):")
display(promedios_asistencia)
```


## ‚è±Ô∏è (8‚Äì10 min) Actividad 5 ‚Äî Mini-reto
**Objetivo:** Aplicar todo lo visto.

**Instrucci√≥n:**
- Selecciona a todos los estudiantes **aprobados** de **Ingenier√≠a** con **asistencia ‚â• 80%**.
- Calcula su **promedio de nota**.


```python

subset = df[(df["Carrera"] == "Ingenier√≠a") & (df["Aprobado"]) & (df["Asistencia_%"] >= 80)]
promedio_sub = subset["Nota_Final"].mean()
print("Promedio de nota (Ingenier√≠a, aprobados, asistencia ‚â• 80%):", round(promedio_sub, 2))
subset.head()
```
