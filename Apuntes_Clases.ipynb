## Métricas de Validación en Machine Learning (valores típicos de aceptación)

> ⚠️ Importante: estos valores son guías generales, no reglas absolutas.
Dependen del dominio, calidad del dataset y tipo de problema.

| Tipo de problema | Métrica | Interpretación | Valor típico aceptable |
|------------------|---------|----------------|-------------------------|
| **Regresión** | **R² (Coef. de determinación)** | % de variabilidad explicada por el modelo | **≥ 0.80** (bueno) |
| | **MAE (Error absoluto promedio)** | Error promedio en la misma unidad que la variable | Mientras más bajo, mejor. Comparar con rango de la variable. |
| | **RMSE (Raíz del error cuadrático)** | Penaliza errores grandes | Idealmente cercano al MAE; bajo respecto al rango de valores. |
| | **MAPE (% error absoluto)** | Error relativo (%) | **< 20%** (aceptable), **< 10%** (bueno) |
| **Clasificación** | **Accuracy** | % de predicciones correctas | **> 0.90** (bueno), pero depende del balance de clases |
| | **Precision** | Qué tan confiables son los positivos predichos | **> 0.80** cuando falsos positivos son críticos |
| | **Recall** | Capacidad de detectar casos positivos | **> 0.80** cuando falsos negativos son críticos |
| | **F1 Score** | Balance entre precision y recall | **> 0.80** (generalmente bueno) |
| | **AUC-ROC** | Capacidad de separar clases | **≥ 0.85** (bueno), **≥ 0.90** (muy bueno) |
| **Clustering** | **Silhouette Score** | Qué tan bien separan los clusters | **0.5 – 1.0** (bueno), **> 0.7** (muy bueno) |
| | **Davies-Bouldin Index** | Menor es mejor | **< 1.0** (bueno) |


## Matriz de Confusión

La **matriz de confusión** es una tabla que nos muestra **cómo se comporta un modelo de clasificación**, comparando lo que el modelo predijo con la realidad.

Se compone de cuatro valores:

|                        | **Predicción: Positivo** | **Predicción: Negativo** |
|------------------------|---------------------------|---------------------------|
| **Real: Positivo**     | **TP** (True Positive)    | **FN** (False Negative)   |
| **Real: Negativo**     | **FP** (False Positive)   | **TN** (True Negative)    |

---


#### ¿Qué significa cada celda?

- **TP (Verdaderos Positivos):**  
  El modelo predijo *positivo* y era *positivo*.  
  → Acierto.

- **TN (Verdaderos Negativos):**  
  El modelo predijo *negativo* y era *negativo*.  
  → Acierto.

- **FP (Falsos Positivos):**  
  El modelo predijo *positivo*, pero era *negativo*.  
  → Error tipo I (“falsa alarma”).

- **FN (Falsos Negativos):**  
  El modelo predijo *negativo*, pero era *positivo*.  
  → Error tipo II (“lo pasó por alto”).

---

#### ¿Por qué es importante?

Porque el **accuracy a veces engaña**.  
Un modelo puede tener alta exactitud, pero cometer errores críticos.

Ejemplo típico en educación:

- **FN:** un estudiante *en riesgo real* que el modelo no detecta  
- **FP:** estudiante *que no está en riesgo*, pero el modelo alerta  

Ambos errores tienen implicancias distintas.

---

#### Métricas que nacen de la matriz

- **Accuracy:** proporción de aciertos  
  \[
  \frac{TP + TN}{TP + TN + FP + FN}
  \]

- **Precision:** de todos los que predije positivos, ¿cuántos eran correctos?  
  \[
  \frac{TP}{TP + FP}
  \]

- **Recall (Sensibilidad):** de todos los positivos reales, ¿cuántos detecté?  
  \[
  \frac{TP}{TP + FN}
  \]

- **F1 Score:** balance entre precision y recall.

---

#### Dicho de otra forma (sencillo)

- **Precision** → “Cuando digo *riesgo*, ¿qué tan seguro estoy?”  
- **Recall** → “De todos los que *sí estaban en riesgo*, ¿cuántos encontré?”  
- **F1** → “¿Qué tan equilibrado está mi modelo?”

## Tabla Resumen de Componentes del Deep Learning

| Categoría | Elementos | Descripción clara y uso típico |
|----------|-----------|--------------------------------|
| **1. Arquitectura del Modelo** | Dense, Flatten, Conv2D, MaxPooling, Dropout, BatchNormalization, LSTM, GRU | Conjunto de capas que definen la estructura y forma del modelo. Determinan cómo fluye la información y qué tipo de problema se puede resolver (tabular, imágenes, secuencias). |
| **2. Funciones de Activación** | ReLU, Sigmoid, Tanh, Softmax, Linear | Introducen no linealidad. ReLU es estándar en capas ocultas; Sigmoid para clasificación binaria; Softmax para multiclase; Linear para regresión. |
| **3. Funciones de Pérdida (Loss Functions)** | categorical_crossentropy, binary_crossentropy, mse, mae, huber | Guían el aprendizaje indicando qué tan equivocada está la red. Para clasificación se usan crossentropy; para regresión, mse o mae. |
| **4. Optimizadores** | SGD, Adam, RMSprop, Adagrad, Adamax | Definen cómo se actualizan los pesos durante el entrenamiento. Adam es el más usado por su estabilidad y rapidez. SGD es más básico pero controlable. |
| **5. Hiperparámetros de Entrenamiento** | epochs, batch_size, learning_rate, momentum, decay, validation_split | Controlan cómo aprende la red: cuántas veces ve los datos (epochs), tamaño de cada lote (batch_size), velocidad de ajuste (learning_rate). |
| **6. Regularización (Anti-Overfitting)** | Dropout, L1/L2 Regularization, EarlyStopping, Data Augmentation | Técnicas para evitar que el modelo memorice. Dropout apaga neuronas durante el entrenamiento; L2 penaliza pesos grandes; EarlyStopping detiene el entrenamiento cuando el modelo deja de mejorar. |
| **7. Métricas de Evaluación** | accuracy, precision, recall, f1-score, auc, mse, mae, rmse, r² | No afectan el entrenamiento, pero permiten evaluar qué tan bien funciona el modelo. Accuracy es para clasificación; mse o mae para regresión. |
| **8. Datos de Entrada y Preprocesamiento** | Normalización, estandarización, one-hot encoding, reshaping, escalamiento | Preparan los datos para que puedan ser entendidos por la red. Normalizar imágenes, escalar características numéricas, convertir etiquetas a one-hot. |
| **9. Proceso de Entrenamiento** | forward pass, backpropagation, gradient descent, minibatch training | Describe cómo la red aprende: predice (forward), calcula error, ajusta pesos con gradiente (backpropagation). Repetido miles de veces. |
| **10. Evaluación y Predicción** | model.evaluate, model.predict, argmax, confusion matrix | Permiten medir el desempeño del modelo y generar predicciones finales. Para clasificación se usa argmax; para regresión se devuelve un valor continuo. |


## Mapa Conceptual de Deep Learning

### 1. Arquitectura del Modelo

#### Capas (dependen del tipo de problema y del tipo de dato)
- **Dense (MLP)** → Para datos tabulares o vectores (no espaciales ni secuenciales)
- **Flatten** → Convierte matrices/imágenes a vectores; útil antes de MLP
- **Conv2D / MaxPooling (CNN)** → Para imágenes; detectan patrones espaciales
- **LSTM / GRU (RNN)** → Para secuencias y series de tiempo; capturan dependencias temporales
- **Dropout** → Técnica de regularización para evitar overfitting (aplicable en cualquier arquitectura)
- **BatchNormalization** → Normaliza activaciones para estabilizar el entrenamiento (aplicable en cualquier arquitectura)

#### Estructura del Modelo (depende del tipo de problema)
- **Número de capas** → Modelos simples usan pocas capas; imágenes/secuencias requieren arquitecturas profundas
- **Neuronas por capa** → Problemas simples usan menos neuronas; problemas complejos requieren más capacidad
- **Funciones de activación** → ReLU (ocultas), Softmax (multiclase), Sigmoid (binaria), Linear (regresión)
- **Forma de entrada (input shape)** → Definida por el tipo de dato: imagen, secuencia, texto o tabular

### 2. Funciones de Activación

- ReLU → capas ocultas
- Sigmoid → clasificación binaria
- Softmax → clasificación multiclase
- Tanh → redes recurrentes
- Linear → regresión

### 3. Funciones de Pérdida (Loss Functions)

#### Clasificación
- categorical_crossentropy
- binary_crossentropy

#### Regresión
- mse (Mean Squared Error)
- mae (Mean Absolute Error)
- huber

### 4. Optimizadores

- Adam (más utilizado)
- SGD (básico, controlable)
- RMSprop
- Adagrad
- Adamax

### 5. Hiperparámetros de Entrenamiento

- epochs
- batch_size
- learning_rate
- momentum
- decay
- validation_split

### 6. Regularización (Anti-Overfitting)

- Dropout
- L1 y L2 (penalización de pesos)
- EarlyStopping
- Data Augmentation (para imágenes)

### 7. Métricas de Evaluación

#### Clasificación
- accuracy  
- precision  
- recall  
- f1-score  
- auc   

#### Regresión
- mse  
- mae  
- rmse  
- r²

### 8. Preparación de Datos

- Normalización  
- Estandarización  
- One-Hot Encoding  
- División train / val / test  
- Escalamiento (MinMax, Z-score)  

### 9. Proceso de Entrenamiento

- Forward Pass  
- Cálculo de la pérdida  
- Backpropagation  
- Ajuste de pesos (Gradient Descent)  
- Iteración por batches y epochs  

### 10. Evaluación y Predicción

- model.evaluate()  
- model.predict()  
- argmax (clasificación)  
- Matriz de confusión  
- Interpretación de errores
