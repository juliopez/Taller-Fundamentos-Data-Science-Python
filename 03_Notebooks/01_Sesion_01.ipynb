## ‚è±Ô∏è 5) Pandas b√°sico (35‚Äì40 min)
DataFrames, inspecci√≥n r√°pida y operaciones comunes.


```python

import pandas as pd

# Crear un DataFrame de ejemplo
data = {"Nombre": ["Ana", "Luis", "Carla", "Diego", "Marta"],
        "Edad": [23, 34, 29, 41, 37],
        "Ciudad": ["Santiago", "Valpara√≠so", "Concepci√≥n", "La Serena", "Rancagua"]}
df = pd.DataFrame(data)
df
```

```python

# Inspecci√≥n r√°pida
display(df.head())
df.info()
df.describe(numeric_only=True)
```

```python

# Operaciones comunes
# Seleccionar columnas
edades = df["Edad"]
print("Edades:", edades)
```

```python

# Filtrar filas (mayores o iguales a 30)
mayores_30 = df[df["Edad"] >= 30]
print("\nMayores o iguales a 30:\n", mayores_30)
```

```python
# Agregar columna calculada
df["Edad+1"] = df["Edad"] + 1
df
```

```python

import numpy as np

# Crear un arreglo
arr = np.array([10, 20, 30, 40, 50])
print("Arreglo:", arr)

# Operaciones matem√°ticas
print("Suma +5:", arr + 5)
print("Promedio:", np.mean(arr))
print("Desviaci√≥n est√°ndar:", np.std(arr))
```

| Caracter√≠stica          | Listas (`list`) | Tuplas (`tuple`)       | Arrays (`numpy`)              |

```python
# (Opci√≥n A) Crear un CSV de ejemplo
import pandas as pd
sample = pd.DataFrame({
    "id": range(1, 11),
    "producto": ["A","B","C","A","B","C","A","B","C","A"],
    "precio": [1000, 1500, 1200, 1100, 1800, 1300, 1150, 1750, 1250, 1050],
    "cantidad": [3, 2, 5, 1, 4, 2, 6, 3, 2, 1]
})
sample.to_csv("sample.csv", index=False)
print("Archivo 'sample.csv' generado en el entorno actual.")
sample.head()
```

```python

# Leer CSV (reemplazar "Ejemplo1.csv" por el archivo real si ya lo subiste)
import pandas as pd
df_csv = pd.read_csv("Ejemplo1.csv")

print("Primeras filas:")
display(df_csv.head())

print("\nInformaci√≥n del dataset:")
df_csv.info()

print("\nEstad√≠sticas descriptivas:")
df_csv.describe(numeric_only=True)
#.describe() got an unexpected keyword argument 'numeric_only'
# En pandas (DataFrame o Series), cuando tienes columnas num√©ricas puede gestionar decimales con .round(X-cantidad decimales)
```

```python

# Ejemplos de preguntas r√°pidas sobre el CSV
total_items = (df_csv["cantidad"]).sum()
ingreso_total = (df_csv["precio"] * df_csv["cantidad"]).sum()

print("Total de √≠tems vendidos:", total_items)
print("Ingreso total (precio*cantidad):", ingreso_total)

# Agrupar por producto
resumen = df_csv.groupby("producto").agg(
    ventas=("cantidad", "sum"),
    ingreso=("precio", lambda s: (s*df_csv.loc[s.index, "cantidad"]).sum())
).reset_index()
resumen
```
